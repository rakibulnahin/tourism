```python
import requests
from PIL import Image
from bs4 import BeautifulSoup
import re
```

```python
re.sub("[0-9]+.\s", "", "2. Burra Homestead, SA")
```

```python
place_names = []
place_texts = ""
img_links = []
website = requests.get("https://www.australiantraveller.com/100-best-views-in-australia")
soup = BeautifulSoup(website.content, 'html.parser')

for i in range(10):
    # print(website)
    section = soup.find('div', id="module-"+str(i+1))
    section = section.find("div", class_ = "eat-section-para")
    # finding names
    module_name = [re.sub("[0-9]+.\s", "", i.text) for i in section.find_all("h3")]
    place_names.extend(module_name)
    # print(len(section.find_all("p")))
    # finding text
    # text = ""

    
    # print(img_links)
    # break
    
    img_count = 0
    for count, para in enumerate(section.find_all("p")):
        # print(para.find_all("img"))
        img = para.find("img")
        if img == None:
            print("image missing of module",i+1, "location number:", count)
        else:
            img_count+=1
            img_links.append((str(i)+"_"+str(count), img["data-src"]))
        print(img_count)
        # for imgs in para.find_all("img"):
        #     img_links.append(imgs["src"])
        # print(len(img_links))
            
        for j in para.find_all(string=True):
            if j == '\n\xa0' or j == '\n':
                # print("yeas")
                continue
            else:
                place_texts += j.replace('\n', " ")

place_texts = place_texts.lower().split("where is it?")[1:]
```

```python
print(len(place_names))
print(len(place_texts))
```

```python
# all the names
for count, i in enumerate(place_names):
    print(count, i)
```

```python
# all texts
for count, i in enumerate(place_texts):
    print(count, i[:50])
```
